# Extracting Zip Dataset

```python
# importing required modules
from zipfile import ZipFile

def extract_zip_dataset(dataset_dir, zip_name, dest_dir):
    # specifying the zip file name
    file_name = os.path.join(dataset_dir, zip_name)

    # opening the zip file in READ mode
    with ZipFile(file_name, 'r') as zip:
        # printing all the contents of the zip file
        # zip.printdir()
        print(len(zip.namelist()))
      
        # extracting all the files
        print('Extracting all the files now...')
        zip.extractall(os.path.join(dataset_dir, dest_dir))
        print('Done!')
        return True
```

```python
# Extract CMFD1
res_cmfd1 = extract_zip_dataset(DATASET_DIR, 'CMFD1.zip', 'CMFD1/')
print(res_cmfd1)
```

```python
# Extract IMFD
res_imfd = extract_zip_dataset(DATASET_DIR, 'IMFD.zip', 'IMFD/')
print(res_imfd)
```

```python
# Extract IMFD1
res_imfd1 = extract_zip_dataset(DATASET_DIR, 'IMFD1.zip', 'IMFD/')
print(res_imfd1)
```

```python
# Debug extracted dir
def debug_extracted_dir(dataset_dir, dir_name):
  dir = os.path.join(dataset_dir, dir_name)
  total_img = 0
  for i in os.listdir(dir):
    img_count = len(os.listdir(os.path.join(dir, i)))
    print(i, img_count)
    total_img += img_count
  print(total_img)
```

```python
# Debug extracted CMFD dir
debug_extracted_dir(DATASET_DIR, 'CMFD/')
```

```python
# Debug extracted CMFD1 dir
debug_extracted_dir(DATASET_DIR, 'CMFD1/')
```
